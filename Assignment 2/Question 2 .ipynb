```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load your specific image \
filename = 'Image of the crop field.png'
img = cv.imread(filename, cv.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Could not find '{filename}'.")
else:
    #Extract the x and y coordinates
    edges = cv.Canny(img, 550, 690)
    indices = np.where(edges != 0)
    x = indices[1] # Column index
    y = indices[0] # Row index

    #Scatter Plot 
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, s=1, c='blue', marker='.')

    plt.gca().invert_yaxis() 

    plt.title('Scatter Plot of Extracted Edge Coordinates')
    plt.xlabel('x (Column Index)')
    plt.ylabel('y (Row Index)')
    plt.grid(True, linestyle='--', alpha=0.6) 
    
    plt.show()
```
