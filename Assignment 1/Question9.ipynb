```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


filename = 'taylor.jpg'
file_path = fr"d:\computer_Vision_Inclass_11_26\{filename}"

img = cv.imread(file_path)

if img is None:
    print(f" Error: Could not find '{filename}'. Check the path")
else:
    print(f" Loaded '{filename}'. Applying sharpening...")

    #  Define Sharpening Kernel 
    # Center = 5, Neighbors = -1. Sum = 1 (keeps brightness the same).
    kernel = np.array([[ 0, -1,  0],
                       [-1,  5, -1],
                       [ 0, -1,  0]])

    # Apply the Filter 
    sharpened = cv.filter2D(img, -1, kernel)

 
    plt.figure(figsize=(15, 8))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title('Original Image\n(Soft Details)')
    plt.axis('off')

    # Sharpened Image
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(sharpened, cv.COLOR_BGR2RGB))
    plt.title('Sharpened Image\n(Look at the Crown & Jewelry!)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
