```python
import cv2 as cv
import matplotlib.pyplot as plt

#  Load the Image
file_path = r"d:\computer_Vision_Inclass_11_26\salt_and_pepper.png"
img = cv.imread(file_path, cv.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Image not found. Check the filename and path")
else:
    print("Image loaded. Applying filters...")

    # 8(a) Apply Gaussian Smoothing
    # Uses a 5x5 kernel.
    #  The noise gets blurry
    gaussian_blur = cv.GaussianBlur(img, (5, 5), 0)

    # 8(b) Apply Median Filtering
    # Uses a 5x5 kernel .
    # The noise is usually completely removed .
    median_filter = cv.medianBlur(img, 5)

    #  Visualization
    plt.figure(figsize=(15, 5))

    #  Original Noisy Image
    plt.subplot(131)
    plt.imshow(img, cmap='gray')
    plt.title('Original (Salt & Pepper Noise)')
    plt.axis('off')

    #  Gaussian Result
    plt.subplot(132)
    plt.imshow(gaussian_blur, cmap='gray')
    plt.title('Gaussian Smoothing\n(Noise is blurred, not removed)')
    plt.axis('off')

    #  Median Result
    plt.subplot(133)
    plt.imshow(median_filter, cmap='gray')
    plt.title('Median Filtering\n(Best for this noise)')
    plt.axis('off')

    plt.show()
```
